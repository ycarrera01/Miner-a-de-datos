```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Modelo de regresiÃ³n logÃ­stica
# Crear y entrenar el modelo
log_reg_model = LogisticRegression()
log_reg_model.fit(X_train, y_train)



```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

# Predecir
y_pred = log_reg_model.predict(X_test)

# Evaluar regresiÃ³n logistica
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Imprimir resultados
print("ðŸ”¹ EvaluaciÃ³n del modelo de regresiÃ³n logÃ­stica:")
print(f"   - Accuracy:  {accuracy:.4f}")
print(f"   - Recall:    {recall:.4f}")
print(f"   - Precision: {precision:.4f}")
print(f"   - F1-Score:  {f1:.4f}")
```

```python
#SVC CON GRID
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Definir el rango de hiperparÃ¡metros
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

# Modelo base SVC
svm_base = SVC()

# Configurar GridSearchCV
grid_search = GridSearchCV(svm_base, param_grid, cv=5, scoring='accuracy')

# Entrenar GridSearchCV
grid_search.fit(X_train, y_train)

# Extraer el mejor modelo encontrado
best_svm_model = grid_search.best_estimator_

# Mostrar mejores hiperparÃ¡metros
print("ðŸ”Ž Mejores hiperparÃ¡metros encontrados:", grid_search.best_params_)


```
